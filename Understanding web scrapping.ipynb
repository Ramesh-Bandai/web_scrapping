# finally Building A function

```python
#Cofee shops in glassgow

import requests
from bs4 import BeautifulSoup
from requests_html import HTMLSession
import pandas as pd
import time

main_list=[]

def extract(url):
    #url='https://www.yell.com/ucs/UcsSearchAction.do?scrambleSeed=540269558&keywords=cafes+%26+coffee+shops&location=glasgow&pageNum=1'
    headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'}
    try:
        session = HTMLSession()
        r = session.get(url,headers=headers)
        
    except requests.exceptions.RequestException as e:
        print(e)
        
    soup=BeautifulSoup(r.content,'html.parser')
    return soup.find_all('div', class_ = 'row businessCapsule--mainRow')



def transform(articles): #return variable assigned to artarticles automatically
    for item in articles:
        Name=item.find('span',class_="businessCapsule--name").text
        address=item.find('span', {'itemprop': 'address'}).text.strip().replace('\n','')
        try:
            website=item.find('a', class_ = 'btn btn-yellow businessCapsule--ctaItem')['href']
            
        except:
            website=''
            
        try:
            tel = item.find('span', class_ = 'business--telephoneNumber').text.strip()
            
        except:
            tel =''
            
        
        business = {
                  'name': Name,
                  'address': address,
                  'website': website,
                  'tel': tel
               }
        
        main_list.append(business)
    return

def load():
    df=pd.DataFrame(main_list)
    df.to_csv('cofeeshopsglassgow.csv',index=False)
        
    
        
#extract() it need to pass url
#extract('https://www.yell.com/ucs/UcsSearchAction.do?scrambleSeed=540269558&keywords=cafes+%26+coffee+shops&location=glasgow&pageNum=1')

for x in range(1,9): # Total 8 pages will be scraped
    print(f'getting page {x}')
    articles=extract(f'https://www.yell.com/ucs/UcsSearchAction.do?scrambleSeed=540269558&keywords=cafes+%26+coffee+shops&location=glasgow&pageNum={x}')
    transform(articles)
    time.sleep(5) # 5 sec
    
    
#print(len(main_list))
load()
print('saved to csv')




            
               
```
